# Problem 1

**Remove duplicate from sorted linked list**

```python
class Node:
    def __init__(self, data):
        self.data = data # Stores data
        self.next = None # Contains reference to the next node if it exists, otherwise None
```

Consider an implementation of a linked list, where each node is created using the given class `Node`. Suppose it has a `head` variable that contains the reference to the first node of the linked list.

You are given a non-empty linked list with `n` nodes, where these nodes are sorted in ascending order of their values. Your task is to remove nodes with duplicate values from the given list.

Write a function **removeDuplicate(head)**, where `head` is a reference to the first node of the sorted linked list, that removes nodes with duplicate values from the given linked list. **The function should not return any value**.

**Sample Input**
```python
1 1 2 2 2 3 3 4 4 4 4 5 7 9 #Input linked list elements
```
**Output**
```python
Output Linked List: 1 2 3 4 5 7 9 #Linked list elements after removing duplicates
```

````python
def removeDuplicate(head):
    current = head
    while current and current.next:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next
````

# Problem 3

**Merge two sorted linked lists**

```python
class Node:
    def __init__(self, data):
        self.data = data # Stores data
        self.next = None # Contains reference to the next node if it exists, otherwise None
```
Consider an implementation of a linked list, where each node is created using the given class `Node`. Suppose it has a `head` variable that contains the reference to the first node of the linked list.

You are given two non-empty linked lists with `n` and `m` nodes, where these nodes are sorted in ascending order of their values. Your task is to merge these two sorted linked lists into one sorted linked list.

Write a function **mergeSortedList(head1, head2)**, where `head1` and `head2` are references to the first nodes of the two sorted linked lists. The function should return the reference to the first node of the merged sorted linked list.

**Sample Input**
```python
1 3 5 7 9 11 #Elements of first linked list
2 4 6 #Elements of second linked list
```

**Output**
```python
Output Linked List: 1 2 3 4 5 6 7 9 11 #Elements of merged sorted linked list
```

````python
def mergeSortedList(head1, head2):
    dummy = Node(0)
    current = dummy
    
    while head1 and head2:
        if head1.data <= head2.data:
            current.next = head1
            head1 = head1.next
        else:
            current.next = head2
            head2 = head2.next
        current = current.next
    
    if head1:
        current.next = head1
    elif head2:
        current.next = head2
    
    return dummy.next
````
